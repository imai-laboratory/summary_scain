```python
# Create tex
import pandas as pd
n_samples = 5
speakers = ["A: ", "B: "]
masked_utterance = "********************************"
third_person = "C: "

for survey_idx in range(n_samples):

    SURVEY_NUMBER = survey_idx + 1
    csv_filename = "./forms/assist{:03}.csv".format(survey_idx)
    assist_data = pd.read_csv(csv_filename)

    head_path = "./tex/head.tex"
    with open(head_path) as head_file:
        head_tex = head_file.read()
    tail_path = "./tex/tail.tex"
    with open(tail_path) as tail_file:
        tail_tex = tail_file.read()

    for idx, row in enumerate(assist_data.itertuples()):
        dialogue_list = row.full_dialogue.split("\n")

        full_body = "\n\n".join(dialogue_list) + "\n"
        full_path = "./forms/assist" + str(SURVEY_NUMBER).zfill(3) + "_" + str(idx).zfill(3) + "full.tex"
        with open(full_path, mode='w') as output_file:
            output_file.write(head_tex)
            output_file.write(full_body)
            output_file.write(tail_tex)

        full_name = "forms/assist" + str(SURVEY_NUMBER).zfill(3) + "_" + str(idx).zfill(3) + "full"
        !ptex2pdf -l {full_name} -output-directory "forms/"

        omit_list = dialogue_list
        for utterance_idx in range(len(omit_list) - 3, len(omit_list) - 1):
            speaker_indicator = ""
            for speaker in speakers:
                if omit_list[utterance_idx].startswith(speaker):
                    speaker_indicator = speaker
            omit_list[utterance_idx] = speaker_indicator + masked_utterance

        omit_body = "\n\n".join(omit_list)
        omit_body = omit_body + "\n\n" + row.assist_utterance
        omit_body = omit_body + "\n\n" + third_person + "\n"

        omit_path = "./forms/assist" + str(SURVEY_NUMBER).zfill(3) + "_" + str(idx).zfill(3) + "omit.tex"
        with open(omit_path, mode='w') as output_file:
            output_file.write(head_tex)
            output_file.write(omit_body)
            output_file.write(tail_tex)

        omit_name = "forms/assist" + str(SURVEY_NUMBER).zfill(3) + "_" + str(idx).zfill(3) + "omit"
        !ptex2pdf -l {omit_name} -output-directory "forms/"
```
